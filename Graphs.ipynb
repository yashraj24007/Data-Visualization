```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
weath.plot.scatter(x='Date'  , y='PM2.5')
```

```python
plt.scatter(x=weath['Date']  , y=weath['PM2.5'])
plt.xlabel('Date')
plt.ylabel('PM2.5')
plt.title('Scatter Plot of Date vs PM2.5')
```

```python
sns.scatterplot(data = weath , x=weath['Date']  , y=weath['PM2.5'])
```

```python
weath.plot.line(x= 'Date' , y = 'PM10')
```

```python
weath_PM10 = weath.groupby("Date")["PM10"].mean()
weath_PM10
```

```python
weath_PM10.plot.line(x = 'Date' , y = 'PM10')
```

```python
weath.plot.bar(x='Date',y='AQI')
```

```python
weath.plot.hist(x='Date',y='NO2', color = 'black')
plt.xlabel('Date')
plt.ylabel('NO2')
plt.title("Histogram of Date vs AQI")
```
